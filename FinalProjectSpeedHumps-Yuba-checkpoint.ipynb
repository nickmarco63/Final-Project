# Introduction

For my portion of the final project, I decided to analyze speed humps in Pittsburgh neighborhoods. Speed humps are made to slow down traffic and make neighborhoods more safe. Neighborhoods with more speed humps will be considered better for this project. 



```python
import pandas as pd

%matplotlib inline
```

```python
speed = pd.read_csv("37b2ac41-ae8e-4de1-8405-157e05dc3640.csv")

speed.head()
```

# Dataset Overview

This dataset contains information about speed humps throughout Pittsburgh.  
The most important column here is 

- the neighborhood where the speed hump is located


```python
speed.columns
```

# Cleaning the Data

There were some parts without the neighborhood sections, so the following code is just meant to remove them


```python
speed = speed.dropna(subset=["neighborhood"])

speed.head()
```

# Counting Speed Humps by Neighborhood


```python
hump_counts = speed.groupby("neighborhood").count()

hump_counts.head()
```

```python
hump_counts = speed["neighborhood"].value_counts()

hump_counts.head(10)
```

# Top Neighborhoods with Most Speed Humps

```python
hump_counts = hump_counts.sort_values(ascending=False)

hump_counts.head(10)
```

# Visualization

This chart shows the top 10 neighborhoods with the most speed humps.

```python
hump_counts.head(10).plot(kind="bar", figsize=(12,6))
```
