# **Best Neighborhood Analysis**

## *Methodology*

Because our data was all over the place for the best nerighborhoods, we needed a way to *restrict* the scope of our criteria to hone in on the best neigborhood. We did this by assuming that we are looking for the **best neighborhood for a new young family**. This means that we had to devise a method to find the best neighborhoods in such a way that our data being scrambled didn't matter. 

To do this we devised a **weighted category system**. We distributed the scale so the weights would **sum to 1**, and determined the magnitude of each scaling based on the perspective of a young family. So we assumed crime incident rate would be the **most important**, with speed humps for pedestrian safety being the **second most important**, population density being the **third**, and 311 requests being the **fourth**. 

Thus the weighting scale was as followed:

- **Crime = 0.40**  
- **Speed Humps = 0.30**  
- **Denisty = 0.20**  
- **311 Requests = 0.10**  

---

## **Caveats**

We do have some caveats with this, typically *less 311 requests would be better*, however because **311 requets are heavily corelated with population**, we assumed **more 311 requests would be better** since *population density was a bigger concern to our hypothetical young family anyways.*

---

## **Methodology for the Data**

I used only neighborhoods that appeared in at least one of our top lists and filled missing values with **0**. These neighborhoods were also the ones that **kept appearing on our tops lists the most.**

---

## **Normalize Data**

Because the magnitude for the datas are so **vastly different** if we applied the weighting system to just these, it would result in the categories like **311 requests domianting the results** since they are all in the **tens of thousands**, even though they are of the least concern from our perspective.

To avoid this, we normalized the data from 0-1 so that each category would be weighted fairly.

---

## **Results**

From here we just apply the weights and sort by the highest score.

```python
import pandas as pd

data = {
    "Neighborhood": [
        "Brookline", "Carrick", "South Side Flats", "South Side Slopes",
        "Squirrel Hill South", "Shadyside", "Bloomfield", "Knoxville"
    ],
    "Crime": [7800, 12000, 19200, 0, 8000, 8800, 8700, 6600],
    "Speed_Humps": [4, 9, 3, 4, 18, 0, 3, 0],
    "Population_Density": [0, 0, 0, 0, 0, 15461, 12060, 12490],
    "Requests_311": [32245, 31017, 29758, 29508, 29226, 0, 0, 0]
}
df = pd.DataFrame(data)
#this aprt just normalizes the data from 0-1
df["Crime_Score"] = 1 - ((df["Crime"] - df["Crime"].min()) / (df["Crime"].max() - df["Crime"].min()))
#crime needs to be reveresed since less crime = better
df["Speed_Score"] = (df["Speed_Humps"] - df["Speed_Humps"].min()) / (df["Speed_Humps"].max() - df["Speed_Humps"].min())
df["Density_Score"] = (df["Population_Density"] - df["Population_Density"].min()) / (df["Population_Density"].max() - df["Population_Density"].min())
df["Requests_Score"] = (df["Requests_311"] - df["Requests_311"].min()) / (df["Requests_311"].max() - df["Requests_311"].min())

#weighting
df["Final_Score"] = (
    df["Crime_Score"] * 0.40 +
    df["Speed_Score"] * 0.30 +
    df["Density_Score"] * 0.20 +
    df["Requests_Score"] * 0.10
)


df = df.sort_values("Final_Score", ascending=False)

print(df[["Neighborhood", "Final_Score"]])

```
